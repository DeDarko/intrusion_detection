```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
```

```python
perplexity_test = np.load("/home/dennis/Documents/Bachelor/runs/run-06/test_perplexities.npy")
perplexity_attacks = np.load("/home/dennis/Documents/Bachelor/runs/run-06/perplexities_attacks.npy")
```

```python
perplexity_test.mean()
```

```python
perplexity_attacks.mean()
```

```python
perplexities = pd.DataFrame(
    {"perplexity": list(perplexity_test) + list(perplexity_attacks), "is_attack"  : [1 for _ in perplexity_test] + [0 for _ in perplexity_attacks]
    }
)
```

```python
perplexities = perplexities.sample(frac=1).reset_index(drop=True)
```

```python
fig, ax = plt.subplots(figsize=(30, 15))
plt.scatter(x=perplexities.index, y=perplexities.perplexity, c=perplexities.is_attack,  s=15, cmap="prism")
```
